```python
import matplotlib.pyplot as plt
import numpy as np

# 
data_size = np.array([50000, 100000, 200000, 300000, 400000, 500000, 600000,
                      800000, 1000000, 2000000, 3000000, 4000000,
                      8000000, 16000000, 32000000, 64000000])
lfu_time = np.array([0.13, 0.23, 0.51, 0.69, 0.94, 1.18, 1.43,
                     1.94, 2.45, 5.55, 8.60, 10.64,
                     23.77, 63.80, 202.97, 699.67])
perfect_time = np.array([0.19, 0.31, 0.68, 0.95, 1.23, 1.40,
                         1.68, 2.36, 2.95, 5.78,
                         7.82, 10.17, 17.52, 28.36,
                         59.41, 115.27])

#  
plt.figure(figsize=(12, 6))
plt.plot(data_size, lfu_time, label='LFU', color='blue', marker='o')
plt.plot(data_size, perfect_time, label='Perfect', color='red', marker='o')

#  
plt.title('The running time of the algorithms depends on the amount of data')
plt.xlabel('The amount of data')
plt.ylabel('Time (s)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 
data_spread = np.array([4, 4.7, 5, 5.14, 5.24, 5.4,
                        5.7, 6, 6.7, 7,
                        7.7, 8, 8.7, 9])
lfu_hits = np.array([489999, 450002, 400667, 312640, 249200, 176950,
                     88751, 44590, 8778, 4510,
                     885, 462, 96, 43])
perfect_hits = np.array([489999, 450002, 400730, 366229, 334066,
                         283982, 183850, 106628,
                         24529, 12282, 2477,
                         1193, 256, 112])

#  
plt.figure(figsize=(12, 6))
plt.plot(data_spread, lfu_hits, label='LFU', color='blue', marker='o')
plt.plot(data_spread, perfect_hits, label='Perfect', color='red', marker='o')

#  
plt.title('The number of hits depending on the data spread')
plt.xlabel('Data spread (logarithmic scale)')
plt.ylabel('Hits')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
